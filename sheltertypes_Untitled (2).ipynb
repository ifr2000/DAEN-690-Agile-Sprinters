```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# File path
file_path = r"C:\Users\manis\Dropbox\My PC (DESKTOP-ERGESDR)\Downloads\School Shelters_P_A_T_P&E_1.xlsx"

# Load the dataset
df = pd.read_excel(file_path)

# Count and Percentage Distribution
category_counts = df['PRI /ALT /TEMP /EMERG '].value_counts()
category_percentages = category_counts / len(df) * 100

# Create a DataFrame for plotting
category_summary = pd.DataFrame({
    'Category': category_counts.index,
    'Counts': category_counts.values,
    'Percentage': category_percentages.values
}).sort_values('Counts', ascending=False)


# Plotting
plt.figure(figsize=(14, 8))
sns.barplot(
    x='Category', y='Counts', data=category_summary, palette=sns.color_palette("Blues_r", len(category_summary)),
    edgecolor='black', linewidth=1.2, width=0.5 , hue = None
)

# Add labels for counts and percentages
for i, row in category_summary.iterrows():
    plt.text(i, row['Counts'] + 1, f"{row['Counts']:,} ({row['Percentage']:.1f}%)",
             ha='center', fontsize=10, color='black', fontweight='bold')

# Add gridlines and styling
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.title('Distribution of Shelter Types by Priority',
          fontsize=16, fontweight='bold', pad=15)
plt.xlabel('Shelter Type', fontsize=14)
plt.ylabel('Counts', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
sns.despine()  # Removes unnecessary spines for a cleaner look
plt.tight_layout()

# Show the plot
plt.show()
```
