```python
#importing all required python libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt #use to visualize dataset vallues
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#code to load infraestructura_Point_Features_Stats dataset values
dataset = pd.read_excel("C:\\Users\\manis\\Dropbox\\My PC (DESKTOP-ERGESDR)\\Desktop\\DAEN 690\\DataVisualization\\Dataset\\Municipality_SLR_Stats_Octubre_3_2023-V02.xlsx")
dataset
```

```python
# Grouping and summarizing the population data
data = dataset.groupby(["name"])['totpop_cy'].sum().sort_values(ascending=False).reset_index(name='Total Population')

# Creating a colorful bar plot using a color palette
g = sns.barplot(x='name', y='Total Population', data=data, palette='coolwarm')

# Rotating the x-axis labels for better readability and updating label size
g.set_xticklabels(g.get_xticklabels(), rotation=70, ha='right', fontsize=10)

# Updating the y-axis label and x-axis label with a professional font size
g.set_xlabel("Municipality", fontsize=12, fontweight='bold')
g.set_ylabel("Total Population", fontsize=12, fontweight='bold')

# Setting the size of the plot
g.get_figure().set_size_inches(14, 6)

# Adding a professional title to the chart
plt.title("Total Population by Municipality", fontsize=14)

# Displaying the plot
plt.show()
```

```python
#identifying total area alloted by municipality for each state and by using this plot we can know which state has covered
#maximum and minimum area
data = dataset.groupby(["name"])['area'].sum().sort_values(ascending=False).reset_index(name='Area')
plt.figure(figsize=(12, 4))
plt.plot(data["name"], data["Area"])
plt.xlabel("State Name")
plt.ylabel("Total Area")
plt.xticks(rotation=70)
plt.show()
```

```python
# Grouping and summarizing the area data by municipality (not state)
data = dataset.groupby(["name"])['area'].sum().sort_values(ascending=False).reset_index(name='Area')

# Setting up the figure size
plt.figure(figsize=(14, 6))

# Creating a bar plot to display total area per municipality
sns.barplot(x='name', y='Area', data=data, palette="coolwarm")

# Adding axis labels with professional formatting
plt.xlabel("Municipality Name", fontsize=12, fontweight='bold')
plt.ylabel("Total Area (in sq km)", fontsize=12, fontweight='bold')

# Rotating the x-axis labels for better readability
plt.xticks(rotation=70, fontsize=10)

# Adding a professional title to the plot
plt.title("Total Area Allocated by Each Municipality", fontsize=16, fontweight='bold', color='darkgreen')

# Displaying the plot
plt.show()
```

```python
def arrangeData(data):
    values = []
    data = data.values
    for i in range(len(data)):
        values.append([data[i,0], 0])
        values.append([data[i,0], data[i,1]])
    values = pd.DataFrame(values, columns=['Municipality Name', 'Expected Land Loss %'])    
    return values
```

```python
#using this analysis we are identifing area expected loss when sea level rise to 10 feet
data = dataset.groupby(["name"])['perc_land_loss_10ft'].sum().reset_index()
data = arrangeData(data)
g = sns.boxplot(x='Municipality Name', y='Expected Land Loss %', data=data)
g.get_figure().set_size_inches(12, 3)
plt.title('Projected Land Loss (%) Due to 10 Feet Sea Level Rise')
plt.xticks(rotation=70)
plt.show()
```
