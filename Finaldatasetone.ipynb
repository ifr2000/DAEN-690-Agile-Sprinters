```python
#importing all required python libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt #use to visualize dataset vallues
import seaborn as sns
```

```python
#code to load infraestructura_Point_Features_Stats dataset values
dataset = pd.read_excel("C:\\Users\\manis\\Dropbox\\My PC (DESKTOP-ERGESDR)\\Desktop\\DAEN 690\\DataVisualization\\Dataset\\infraestructura_Point_Features_Stats.xlsx")
dataset
```

```python
#finding sum of missing values
dataset.isnull().sum()
```

```python
#describing dataset to know distribution of values ranges in each column
dataset.describe()
```

```python
#visualizing graph of different layers with different sea feets level
#in graph x-axis represents layer names and y-axis represents counts of feets
#each different line represents feet type as 10, 7, 3 or 1
data = dataset.groupby(["layer_name", "level"]).size().sort_values(ascending=False).reset_index(name='count')
g = sns.catplot(x="layer_name", y="count", hue='level', data=data, kind='point', height=3, aspect=1.9)
g.set_xticklabels(rotation=90)
g.set_axis_labels("Infrastructure Type", "Count of Structures Affected")
plt.title("Impact of Expected Sea Level Rise on Infrastructure by Type and Level")
plt.show()
```

```python
df = pd.read_csv("C:\\Users\\manis\\Dropbox\\My PC (DESKTOP-ERGESDR)\\Desktop\\DAEN 690\\DataVisualization\\Dataset\\infraestructura_Point_Features_Stats.xlsx")

# Check for missing values and handle them (optional)
df.dropna(subset=['layer', 'infrastructure_type'], inplace=True)

# Set up the plot style for professionalism
sns.set(style="whitegrid")

# Create a box plot to visualize the variation in sea level rise ('layer') across infrastructure types
plt.figure(figsize=(10, 6))
sns.boxplot(x='infrastructure_type', y='layer', data=df, palette='coolwarm')

# Add titles and labels for clarity
plt.title('Expected Sea Level Rise by Infrastructure Type', fontsize=16)
plt.xlabel('Infrastructure Type', fontsize=14)
plt.ylabel('Expected Sea Level Rise (Layer)', fontsize=14)
```
